# Data Absenteeism Data Cleaning 


## using -
* get_dummies
* pd.to_datetime
* df.apply(lambda)
* df['columns'].map({})

### Things to do:
* Drop the ‘ID’ column
* Split the reasons for absence into multiple dummy variables, and then group them in the following way:
** Group 1: Columns 1 to 14
** Group 2: Columns 15, 16, and 17
** Group 3: Columns 18, 19, 20, and 21
** Group 4: Columns 22 to 28
* After you’ve done that, don’t forget to drop the ‘Reason for Absence’ column.
* Extract the month value and the day of the week from the ‘Date’ column. Then, drop the ‘Date’ column as well.
* Turn the data from the ‘Education’ column into binary data, by mapping the value of 0 to the value of 1, and the value of 1 to the rest of the values found in this column

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df=pd.read_csv('Absenteeism-data.csv')
df.shape
```

```python
df.head()
```

### dropping the id column

```python
df.drop('ID',axis=1,inplace=True)
```

```python
df.head()
```

### grouping the data in reason for absence col

```python
reasons=pd.get_dummies(df['Reason for Absence'])
reasons
```

```python
reasons['check']=reasons.sum(axis=1)
```

```python
reasons
```

```python
reasons.check.sum()
```

```python
reasons.shape[0]
```

```python
reasons.drop('check',axis=1,inplace=True)
```

```python
reason_1 = reasons.loc[:, 1:14].max(axis=1)
reason_2 = reasons.loc[:, 15:17].max(axis=1)
reason_3 = reasons.loc[:, 18:21].max(axis=1)
reason_4 = reasons.loc[:, 22:].max(axis=1)
```

```python
df2=pd.concat([df,reason_1,reason_2,reason_3,reason_4],axis=1)
```

```python
df2.head()
```

```python
df2.drop('Reason for Absence',axis=1,inplace=True)
```

```python
df2.columns.tolist()
```

```python
df2.columns=['Date',
        'Transportation Expense',
        'Distance to Work',
        'Age',
        'Daily Work Load Average',
        'Body Mass Index',
        'Education',
        'Children',
        'Pets',
        'Absenteeism Time in Hours',
        'Reason_1',
        'Reason_2',
        'Reason_3',
        'Reason_4']
```

```python
df2.head()
```

```python
df2=df2[['Reason_1',
        'Reason_2',
        'Reason_3',
        'Reason_4','Date',
        'Transportation Expense',
        'Distance to Work',
        'Age',
        'Daily Work Load Average',
        'Body Mass Index',
        'Education',
        'Children',
        'Pets',
        'Absenteeism Time in Hours']]
```

```python
df2.head()
```

### extracting months and day of the week

```python
df2.dtypes.Date
```

```python
df2['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y')
```

```python
df2['Month']=df2.Date.apply(lambda x: x.month)
```

```python
df2['Month']
```

```python
df2['Day_of_the_week']=df2.Date.apply(lambda x: x.weekday()) 
```

```python
df2['Day_of_the_week']
```

### mapping education column

```python
df['Education']
```

```python
df2['Education']=df2['Education'].map({1:0,2:1,3:1,4:1})
```

```python
df2
```

```python
final=df2.drop('Date',axis=1)
```

```python
final.to_csv('Visualise_Absenteeism.csv')
```
